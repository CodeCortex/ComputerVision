```python
import cv2 as c
import pyautogui as p
import numpy as np

# Get screen resolution
screen_width, screen_height = p.size()

# Get filename from user
# fn = input("Please enter file name (e.g., output.avi): ")
fn= "output1.avi"

# Set recording parameters
fps = 30.0  # Reduced FPS for better performance
fourcc = c.VideoWriter_fourcc(*'MJPG')  # Try 'XVID', 'MP4V'
output = c.VideoWriter(fn, fourcc, fps, (screen_width, screen_height))

# Check if video writer is working
if not output.isOpened():
    print("Error: Video file not opened for writing!")
    exit()

# Create recording window
c.namedWindow("Live_Recording", c.WINDOW_NORMAL)
c.resizeWindow("Live_Recording", (600, 400))

# Start recording loop
while True:
    img = p.screenshot()  # Capture screenshot
    f = np.array(img)  # Convert to NumPy array
    f = c.cvtColor(f, c.COLOR_RGB2BGR)  # Convert RGB to BGR

    output.write(f)  # Write frame to video
    c.imshow("Live_Recording", f)  # Display recording

    # Stop recording when 'q' is pressed
    if c.waitKey(1) & 0xFF == ord("q"):
        break

# Release resources
output.release()
c.destroyAllWindows()

print(f"Recording saved as: {fn}")
```

#save images upto 10 6 frames

```python


# Load the video file
video_path = "/Users/roshanjaiswal/Downloads/localVideo.MOV"
vidcap = c.VideoCapture(video_path)

# Create a folder to store frames
save_folder = "frames1"
if not os.path.exists(save_folder):
    os.makedirs(save_folder)

count = 0
saved_frames = 0  # Counter to track saved frames

while True:
    ret, image = vidcap.read()  # Read frame inside the loop

    if not ret:
        print("ERROR: Video is not being read or has ended.")
        break  # Exit the loop if video cannot be read

    count += 1

    # Save only 10 frames
    if saved_frames < 6:
        frame_filename = f"{save_folder}/frame_{saved_frames + 1}.jpg"
        c.imwrite(frame_filename, image)  # Save the frame as an image
        print(f"Saved {frame_filename}")
        saved_frames += 1

    c.imshow("res", image)  # Show video frame

    # Exit when 'q' is pressed or when 10 frames are saved
    if c.waitKey(1) & 0xFF == ord("q") or saved_frames >= 10:
        break

# Release resources properly
vidcap.release()
c.destroyAllWindows()

print(f"Successfully saved {saved_frames} frames in the '{save_folder}' folder.")
```

```python
# Load the video file
video_path = "/Users/roshanjaiswal/Downloads/localVideo.MOV"
vidcap = c.VideoCapture(video_path)

# Create a folder to store frames
save_folder = "frames2"
if not os.path.exists(save_folder):
    os.makedirs(save_folder)

count = 0
saved_frames = 0  # Counter to track saved frames

while True:
    ret, image = vidcap.read()  # Read frame inside the loop

    if not ret:
        print("ERROR: Video is not being read or has ended.")
        break  # Exit the loop if video cannot be read

    count += 1

    # Save only 10 frames
    if saved_frames < 2:
        frame_filename = f"{save_folder}/frame_{saved_frames + 1}.jpg"
        c.imwrite(frame_filename, image)  # Save the frame as an image
        vidCap.set(c.CAP_PROP_POS_MSEC, (count*10000)) #Moves video to a specific timestamp (in milliseconds).

        print(f"Saved {frame_filename}")
        saved_frames += 1

    c.imshow("res", image)  # Show video frame

    # Wait for key press
    key = c.waitKey(1) & 0xFF  

    # Exit when 'q', 'Q', or 'space' is pressed or when 10 frames are saved
    if key in [ord("q"), ord("Q"), ord(" ")]:  # Checking multiple keys
        break

    if saved_frames >= 2:  
        break

# Release resources properly
vidcap.release()
c.destroyAllWindows()

print(f"Successfully saved {saved_frames} frames in the '{save_folder}' folder.")
```
