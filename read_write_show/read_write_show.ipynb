```python
# Read the image (-1 keeps the original color depth)
greyImage = cv2.imread("../assets/CodeCortex.jpeg", -1)
print("greyImage shape",greyImage.shape)  # Print image dimensions

# Convert BGR to RGB (Matplotlib uses RGB format)
convertGreyImage = cv2.cvtColor(greyImage, cv2.COLOR_BGR2RGB)
# convertGreyImage= cv2.resize(convertGreyImage, (1280, 6000));
print("convertGreyImage= ",convertGreyImage.shape);

# Create a figure with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
# plt.figure(figsize=(5, 2))

# Show original image
axes[0].imshow(greyImage)
axes[0].set_title("Original Image (BGR)")
axes[0].axis("off")  # Hide axes

# Show converted image
axes[1].imshow(convertGreyImage)
axes[1].set_title("Converted Image (RGB)")
axes[1].axis("off")  # Hide axes

# Show both images together
plt.show()
```

```python
image1= cv2.imread(path,0);
image2= cv2.resize(image1, (560,700));
print(image2);
print("image shape after resize", image2.shape)

# image2= cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

plt.imshow(image2,cmap='gray');
plt.title("Grey image")
plt.axis("off");
plt.show()
```
